Data description: 
- The data set ‘players.csv’ was a table created using a Minecraft server managed by Frank Wood (project lead), which collects data using the information given upon registration (name, age, gender and experience, hashed email) and collects the time played while using the server. 
- There are a total of 7 data-containing columns, including 3 nominal categorical variables (experience, name and gender), 1 binary variable (subscribed), and one identifier (hashed email), and 2 quantitative variables (played_hours, age).
- The dataset contains two columns with no data. Including individual ID and organization name, this information will need to be filtered out to create a tidy data set, as they are unusable. 
Each variable represents a player's information, which can be used to make predictions on players' actions throughout the game
- There are 197 rows of observations, which include all variable information, excluding individual ID and organization name. 


```python
import altair as alt
import pandas as pd
players=pd.read_csv('players.csv').drop(columns=['hashedEmail','individualId','organizationName'])
players
```

```python
players_age=alt.Chart(players, title= 'Age vs Hours contributed').mark_bar().encode(
    x=alt.X('age:Q').title('Age of Players').bin(maxbins=10).scale(zero=False),
    y=alt.Y ('played_hours:Q').title('Hours Contributed (Hours)').scale(zero=False),
)
players_age
```

```python
players_gender=alt.Chart(players, title= 'Gender vs Hours Contributed').mark_bar().encode(
    x=alt.X('gender').title('Gender of Players').scale(zero=False),
    y=alt.Y ('played_hours').title('Hours Contributed (Hours)').scale(zero=False),
)
players_gender
```
